# Project Title: Predict a Poisonous vs. Edible Mushroom using Gradient Boosting Machine Learning algorithm

## Project Background

As per WildFoodUK (https://www.wildfooduk.com/articles/how-to-tell-the-difference-between-poisonous-and-edible-mushrooms/), it is extremely difficult to find the differences between edible and poisonous mushrooms. The general rule is that if you see other animals eating a mushroom, then that should be safe for humans. The best way is to identify the mushroom type as that would help in deciding if it is safe or not. You can check out for the following features to help identify a mushroom:

- The texture of the mushroom
- Size, shape and colour of the cap
- Smell
- Where is the mushroom growing?

There is a nice document listing a variety of edible mushrooms we found and we are sharing it for your benefit:

- http://www.bayareamushrooms.org/images/pdf_documents/EDIBLE%20VS%20POISONOUS%20MUSHROOMS.pdf

## Project Methodology

# Problem Identification

**Problem Hypothesis:** Predict edible vs poisonouse mushroom using gradient boosting classification machine learning algorithms.

**Context for this Problem:** As mushroom hunting is a popular hobby for many mushroom-lovers, it is important to ensure that there is a proper guide made available to those who want to eat mushrooms.

# Data Wrangling

## Identification and Import of Data Set

In order to support the scientific studies, UCI Machine Learning Repository (https://archive.ics.uci.edu/ml/datasets/mushroom) has made this data set open source for machine learning and other scientific studies. I have downloaded it from there and it is now available on my desktop for my project.

**About Dataset:**

This dataset includes descriptions of hypothetical samples corresponding to 23 species of gilled mushrooms in the Agaricus and Lepiota Family Mushroom drawn from The Audubon Society Field Guide to North American Mushrooms (1981). Each species is identified as definitely edible, definitely poisonous, or of unknown edibility and not recommended. This latter class was combined with the poisonous one. The Guide clearly states that there is no simple rule for determining the edibility of a mushroom; no rule like "leaflets three, let it be'' for Poisonous Oak and Ivy.

# Pre-Processing and Training Data Development

## Encoding Categorical Data to Numerical Data

```python
# Using pandas get dummy function to convert for X features only. So splitting into X and y first
X = df.drop('class',axis=1)
y = df['class']

X = pd.get_dummies(X, drop_first=True)
```

## Splitting into Training and Testing Data Set

```python
# Importing required packages
from sklearn.model_selection import train_test_split
```

```python
# Splitting into training and testing (15%) set
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.15, random_state=101)
```

# Machine Learning Modeling

## Gradient Boosting Classifier Modeling with GridSearchCV

```python
# Importing required packages
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, plot_confusion_matrix, accuracy_score
```

```python
# Setting up grid parameters
param_grid = {'n_estimators':[50,100],
             'learning_rate':[0.05,0.1,0.2],
             'max_depth':[3,4,5]}
```

```python
# grid modeling
gb_model = GradientBoostingClassifier()
grid = GridSearchCV(gb_model,param_grid)
grid.fit(X_train,y_train)

preds_grid = grid.predict(X_test)
```

```python
# printing classification report for the grid model
print("Classification Report for Grid Model:\n", classification_report(y_test,preds_grid))
```

```python
# plotting confusion matrix for the grid model
plot_confusion_matrix(grid,X_test,y_test);
```

```python
# Finding best parameters
grid.best_params_
```

```python
# Checking for the feature importances
feat_imp = grid.best_estimator_.feature_importances_
# Creating a dataframe for this
feature_imp = pd.DataFrame(index=X.columns, data=feat_imp, columns=['Importance'])
feature_imp = feature_imp.sort_values('Importance', ascending=False)
feature_imp
```

```python
# Filtering to only those features which are greater than 0.0005
feats = feature_imp[feature_imp['Importance']>0.0005]
feats = feats.sort_values('Importance', ascending=False)
```

```python
# plotting for important features now
plt.figure(figsize=(8,4), dpi=200)
sns.barplot(x=feats.index, data=feats, y='Importance')
plt.title("Bar Plot showing Feature Importance")
plt.xticks(rotation=90)
plt.show();
```

## Gradient Boosting Classifier Modeling with Best Hyperparameters

```python
# Modeling with the best hyperparameters
gbmodel_best = GradientBoostingClassifier(n_estimators=100,learning_rate=0.05,max_depth=4)

gbmodel_best.fit(X_train,y_train)
preds_best = gbmodel_best.predict(X_test)
```

```python
# printing classification report for the best hyperparameters model
print("Classification Report with Best Hyperparameters:\n", classification_report(y_test,preds_best))
```

```python
# plotting confusion matrix for the best hyperparameters model
plot_confusion_matrix(gbmodel_best,X_test,y_test);
```

```python
# Checking for the feature importances
feats = pd.DataFrame(index=X.columns, data=gbmodel_best.feature_importances_, columns=['Importance'])
```

```python
# Sorting feature values
feats_sorts = feats.sort_values('Importance', ascending=False)

# Filtering with 0.0005

feats_important = feats_sorts[feats_sorts['Importance']>0.0005]
```

```python
# plotting for important features now
plt.figure(figsize=(6,2), dpi=200)
sns.barplot(x=feats_important.index, data=feats_important, y='Importance')
plt.title("Bar Plot showing Feature Importance")
plt.xticks(rotation=90)
plt.show();
```
