# Can We Predict if a Cell is Malignant or Benign with Machine Learning?

## Introduction

Machine learning applications in cancer prognosis and prediction is a constantly evolving trend. As machine learning techniques are getting better, many scientists are predicting that the machine learning is the future of cancer prediction.

As per this [scientific journal](https://journals.sagepub.com/doi/pdf/10.1177/1748301818756225), breast cancer is the second most leading cancer in women compared to all other cancers. 

In order to support the research and prepare future machine learning scientists, [Center for Machine Learning and Intelligent Systems](https://archive.ics.uci.edu/ml/datasets/breast+cancer) has made the breast cancer data publicly available for research and scientific studies.

## Problem Identification and Data Features Description

The human body has about 100 trillion cells and usually those cells behave in a certain way. However, occasionally, one of these 100 trillion cells, behave in a different way and keeps dividing and pushing other cells around it out of the way. That cell stops observing the rules of the tissue within which it is located and begins to move out of its normal position and starts invading into the tissues around it and sometimes entering the bloodstream and becoming is called a metastasis.


[Image Source Credit: cancer.gov](https://www.cancer.gov/about-cancer/understanding/what-is-cancer)

In summary, as we grow older,throughout a lifetime, we go through this knid of situation where a particular kind of gene is mutated where the protein that it makes is abnormal and drives the cell to behave in a different way that we call cancer. This is what Dr. WIlliam H. Wolberg was observing, who put together this dataset.

**Attribute Information:**

- Sample code number: id number 
- Clump Thickness: 1 - 10
- Uniformity of Cell Size: 1 - 10 
- Uniformity of Cell Shape: 1 - 10 
- Marginal Adhesion: 1 - 10 
- Single Epithelial Cell Size: 1 - 10 
- Bare Nuclei: 1 - 10 
- Bland Chromatin: 1 - 10 
- Normal Nucleoli: 1 - 10 
- Mitoses: 1 - 10 
- Class: (2 for benign, 4 for malignant)

**Malignant** ==> Cancerous

**Benign** ==> Not Cancerous (Healthy)



## Library and Packages Imports

```python
# Importing required packages and libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from scipy.stats import zscore 
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

%matplotlib inline
import warnings
warnings.simplefilter(action='ignore')
```

## Loading the Dataset

```python
# Reading the data from the local file location
data = pd.read_csv('breastCancer.csv')

# Checking data rows and columns
print("Number of rows are: ",data.shape[0])
print("Number of columns are: ",data.shape[1])
```

# Data Wrangling

```python
data.head()
```

From the above dataset, 'class' is the target variable which is labeled as **'2' for benign and '4' for malignant cancer**. So this is a classification problem and I will be using **classification algorithms** such as K-Nearest Neighbors (KNN) and Support Vector Machine (SVM).

```python
# Plotting countplot() for the target variable 'class'
sns.countplot(data['class'])
plt.title("Count Plot for the Target Variable ('Class')")
plt.show()
```

```python
# Printing the number of values for label 2 and 4
data['class'].value_counts()
```

This means that there are 458 benign labels and 241 malignant labels.

```python
# Checking the data types
data.dtypes
```

All the columns here are integer type and only 'bare_nucleoli' column is 'object' type. I am going to investigate further why this is object type because visually it shows the numbers as integers.

```python
# Checking object column
data['bare_nucleoli'].unique()
```

From the above it shows that everything is normal except for the '?' which is responsible for making this column as object. I am going to check how many records contain '?'.

```python
# Checking for the ? records
data[data['bare_nucleoli']=='?'].sum()
```

```python
# Checking for how many records are there with '?' in this column
digits_in_bare_nucleoli = pd.DataFrame(data['bare_nucleoli'].str.isdigit())
print("Number of rows with '?' in the column are:", (data.shape[0] - digits_in_bare_nucleoli.sum()))
```

```python
# Replacing these '?' with NaN
data = data.replace('?', np.nan)
```

## Missing Values

```python
# Checking for the missing values
data.isnull().sum()
```

Since I replaced '?' values in the 'bare_nucleoli' column with the NaN, it is expected that there will be 16 missing values in that column. I am going to replace these NaN values with the median values rather than dropping it because it is about the health data and dropping it may not be a good option here.

```python
# Imputing with the median values
data = data.fillna(data.median())
```

```python
# Changing the data type for the column 'bare_nucleoli' to integer
data['bare_nucleoli'] = data['bare_nucleoli'].astype('int64')
data.dtypes
```

# Exploratory Data Analysis

Since 'id' column is not adding any value to the data analysis, I am going to drop it.

```python
# Dropping 'id' column
data.drop('id', axis=1, inplace=True)
data.head()
```

```python
# Checking for the describe() for the dataset
data.describe().T
```

# Pre-Processing and Training Data Development

```python
# Checking data again
data.head()
```

```python
# Checking data types
data.dtypes
```

```python
# Dividing data into X and y
X = data.drop('class', axis=1)
y = data['class']
```

```python
# Dividing data into training and testing dataset using 70:30 split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

# Modeling

## K-Nearest Neighbors (KNN) Algorithm

```python
# Developing KNN-model with k-value as 6
knn = KNeighborsClassifier(n_neighbors=6, weights='distance')
knn.fit(X_train, y_train)
y_pred1 = knn.predict(X_test)
```

```python
# Printing predicted values
y_pred1
```

```python
print("KNN model score is: {:0.2f}%".format(knn.score(X_test, y_test) * 100))
```

```python
# Building confusion matrix for the KNN-model predictions
cm_knn = metrics.confusion_matrix(y_test, y_pred1, labels=[2, 4])
df_cmknn = pd.DataFrame(cm_knn, index=['2', '4'], columns=['predict B', 'predict M'])
```

```python
# Building heatmap for the confusion matrix
plt.figure(figsize=(4,4))
sns.heatmap(df_cmknn, annot=True, cmap='viridis', cbar=False, fmt='.0f')
plt.title("Confusion Matrix for the KNN-Model")
plt.show()
```

The KNN model score is 97.62%. There are only 1 False Positive (Malignant) and 4 False Negative (Benign) predictions.

### KNN Hyper-Parameters Tuning

```python
# Running KNN model for different K-values
scores=[]
for k in range(1,50):
    model = KNeighborsClassifier(n_neighbors=k)
    cv_score = cross_val_score(model,X_train, y_train, cv=10)
    scores.append(cv_score.mean())
```

```python
# Plotting different values of k and its score
plt.figure(figsize=(12,8))
plt.plot(range(1,50), scores, color='blue', linestyle='dashed', marker='o', markerfacecolor='red')
plt.xlabel('K-Value')
plt.ylabel('KNN Model Score')
plt.title("KNN Model Score for different K-Values")
plt.show()
```

Based on the above the best k value is 6 which we have already used in the earlier model.

## Support Vector Machine (SVM) Algorithm

```python
# Developing SVM model
svc = SVC(gamma=0.01, C=0.1)
svc.fit(X_train, y_train)
y_pred2 = svc.predict(X_test)
```

```python
y_pred2
```

```python
# Printing the SVM model score
print("SVM model score is {:0.2f}%".format(svc.score(X_test, y_test) * 100))
```

```python
# Confusion matrix for SVM model
cm_svm = metrics.confusion_matrix(y_test, y_pred2, labels=[2,4])
df_cmsvm = pd.DataFrame(cm_svm, index=['2', '4'], columns=['predict B', 'predict M'])
```

```python
# Building heatmap for the confusion matrix
plt.figure(figsize=(4,4))
sns.heatmap(df_cmsvm, annot=True, cmap='viridis', cbar=False, fmt='.0f')
plt.title("Confusion Matrix for the SVM-Model")
plt.show()
```

The SVM model score is 97.62%. There are only 3 False Positive (Malignant) and 2 False Negative (Benign) predictions.

### SVM Hyper-Parameters Tuning

```python
# Using GridSearchCV for the hyper parameter tuning
param_grid = {'C': [0.1, 1, 3, 10, 100],
             'gamma': [1, 0.01, 0.001, 0.025, 0.0001]
             }
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=3)

# Fitting the model for the grid search
grid.fit(X_train, y_train)
```

```python
# Finding the best parameters
print("Best parameters for SVM model are:", grid.best_params_)
```

```python
# Finding the best estimators
print("Best estimators for SVM model are:", grid.best_estimator_)
```
