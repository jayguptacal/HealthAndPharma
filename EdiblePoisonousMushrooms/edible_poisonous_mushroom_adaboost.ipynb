# Project Title: Predict a Poisonous vs. Edible Mushroom using AdaBoost Machine Learning algorithm

## Project Background

As per WildFoodUK (https://www.wildfooduk.com/articles/how-to-tell-the-difference-between-poisonous-and-edible-mushrooms/), it is extremely difficult to find the differences between edible and poisonous mushrooms. The general rule is that if you see other animals eating a mushroom, then that should be safe for humans. The best way is to identify the mushroom type as that would help in deciding if it is safe or not. You can check out for the following features to help identify a mushroom:

- The texture of the mushroom
- Size, shape and colour of the cap
- Smell
- Where is the mushroom growing?

There is a nice document listing a variety of edible mushrooms we found and we are sharing it for your benefit:

- http://www.bayareamushrooms.org/images/pdf_documents/EDIBLE%20VS%20POISONOUS%20MUSHROOMS.pdf

## Project Methodology

# Problem Identification

**Problem Hypothesis:** Predict edible vs poisonouse mushroom using adaboost classification machine learning algorithms.

**Context for this Problem:** As mushroom hunting is a popular hobby for many mushroom-lovers, it is important to ensure that there is a proper guide made available to those who want to eat mushrooms.

# Data Wrangling

## Identification and Import of Data Set

In order to support the scientific studies, UCI Machine Learning Repository (https://archive.ics.uci.edu/ml/datasets/mushroom) has made this data set open source for machine learning and other scientific studies. I have downloaded it from there and it is now available on my desktop for my project.

**About Dataset:**

This dataset includes descriptions of hypothetical samples corresponding to 23 species of gilled mushrooms in the Agaricus and Lepiota Family Mushroom drawn from The Audubon Society Field Guide to North American Mushrooms (1981). Each species is identified as definitely edible, definitely poisonous, or of unknown edibility and not recommended. This latter class was combined with the poisonous one. The Guide clearly states that there is no simple rule for determining the edibility of a mushroom; no rule like "leaflets three, let it be'' for Poisonous Oak and Ivy.

# Pre-Processing and Training Data Development

## Encoding Categorical Data to Numerical Data

```python
# Using pandas get dummy function to convert for X features only. So splitting into X and y first
X = df.drop('class',axis=1)
y = df['class']

X = pd.get_dummies(X, drop_first=True)
```

## Splitting into Training and Testing Data Set

```python
# Importing required packages
from sklearn.model_selection import train_test_split
```

```python
# Splitting into training and testing (15%) set

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.15, random_state=101)
```

# Machine Learning Modeling

## AdaBoost Classifier Modeling with default DecisionTreeClassifier as Base Estimator and Single Stump as n_estimators=1

```python
# Importing required packages
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report, plot_confusion_matrix, accuracy_score
```

```python
# By putting n_estimators=1, model will push to find the best feature to model on
model_1 = AdaBoostClassifier(n_estimators=1)
model_1.fit(X_train,y_train)

preds_1 = model_1.predict(X_test)
```

```python
# printing classification report
print("Classification Report with n_estimators=1:\n", classification_report(y_test,preds_1))
```

```python
# plotting confusion matrix
plot_confusion_matrix(model_1,X_test,y_test);
```

Even with 1 feature, the classification performance is very impressive. Now I am going to try to find out which feature it is picking on to classify the dataset.

```python
# finding the featue AdaBoost is picking for modeling
model_1.feature_importances_.argmax()
```

```python
# Feature at 22 
X.columns[22]
```

So AdaBoost is picking no odor is one of the best classification criteria.

```python
# To dive into this further, we can do countplot
sns.countplot(data=df, x='odor', hue='class');
```

**Here it shows that 'no odor' is one of the most common features for the dataset.**

## AdaBoost Classifier Modeing for Different Number of Features to Find the Right Number of Features Required for the Modeling

```python
# Since there are a large number of features in the dataset, creating a error vs. number of features loop
len(X.columns)
```

```python
# for loop for the error vs features testing
error_rates = []

for n in range(1,96):
    model = AdaBoostClassifier(n_estimators=n)
    model.fit(X_train,y_train)
    preds = model.predict(X_test)
    
    err = 1-accuracy_score(y_test,preds)
    error_rates.append(err)
```

```python
# plotting number of features vs error rates
plt.plot(range(1,96),error_rates)
```

Based on the above, it seems that n_estimators=18 is the right number for the features to be considered for the AdaBoost Classification modeing.

## AdaBoost Classification Modeling with n_estimators=18

```python
# running n_estimators=18 model
model_18 = AdaBoostClassifier(n_estimators=18)
model_18.fit(X_train,y_train)
preds_18 = model_18.predict(X_test)
```

```python
# printing classification report
print("Classification Report with n_estimators=18:\n", classification_report(y_test,preds_18))
```

```python
# plotting confusion matrix
plot_confusion_matrix(model_18,X_test,y_test);
```

This is a phenomenal output with 100% accuracy. Now checking for what are those 18 features for this modeling?

```python
# Checking for features
feats = pd.DataFrame(index=X.columns, data=model_18.feature_importances_, columns=['Importance'])
```

```python
# sorting the feature values
feats_sort = feats.sort_values('Importance', ascending=False)
feats_sort
```

```python
# Since there are some features where importance score is zero, so filtering where it is higher than zero
feats_imp = feats_sort[feats_sort['Importance']>0]
```

```python
# plotting for important features now
plt.figure(figsize=(8,4), dpi=200)
sns.barplot(x=feats_imp.index, data=feats_imp, y='Importance')
plt.title("Bar Plot showing Feature Importance")
plt.xticks(rotation=90)
plt.show();
```

```python
# So here is the list of important features to be used while deciding edible vs poisonous mushrooms

feats_imp.index
```
